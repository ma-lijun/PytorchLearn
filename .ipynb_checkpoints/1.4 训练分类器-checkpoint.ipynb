#### 2.1 加载并标准化CIFAR10

```python
# 使用torchvision, 加载CIFAR10
import torch
import torchvision
import torchvision.transforms as transforms
```

> TorchVision数据集的输出是[0, 1]范围的PILImage图像。我们将它们转换为归一化范围[-1, 1]的张量。

```python
transform = transforms.Compose(
    [transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))
    ]
)
```

```python
transform
```

```python
trainset = torchvision.datasets.CIFAR10(root='../ml/data', train=True, download=False, transform=transform)
```

```python
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4, shuffle=True, num_workers=0)
```

```python
testset = torchvision.datasets.CIFAR10(root='../ml/data', train=False, download=False, transform=transform)
```

```python
testloader = torch.utils.data.DataLoader(testset, batch_size=4, shuffle=False, num_workers=0)
```

```python
classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

### 2. 定义卷积神经网络

> 之前从“神经网络”部分复制神经网络， 然后对其进行修改以获得3通道图像（而不是定义的1通道图像）

```python
import torch.nn as nn 
import torch.nn.functional as F
```

```python
class Net(nn.Module):
    
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6,16,5)
        self.fc1 = nn.Linear(16 * 5 *5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)
        
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 16*5*5)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
net = Net()
```

```python
net
```

### 3. 定义损失函数和优化器

> 此处我们使用分类交叉熵损失和带有动量的SGD 。

```python
import torch.optim as optim
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

### 4. 训练网络

> 这是事情开始变得有趣的时候。我们只需要遍历数据迭代器， 然后将输入馈送到网络并进行优化即可。

```python
for epoch in range(2):
    running_loss = 0.0 
    for i, data in enumerate(trainloader, 0):
        # 获取输入；此处data is list of [inputs, labels]
        inputs, labels = data
        
        # zero the parameter gradients
        optimizer.zero_grad()
        
        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        # print statistics
        running_loss += loss.item()
        # print every 2000 mini-batches
        if i%2000 == 1999:
            print('[%d, %5d] loss %.3f' % (epoch+1, i+1, running_loss/2000))
            running_loss = 0.0
print('Finished training')
            
```

> 快速保存我们的训练过的模型

```python
PATH = './cifar_net.pth'
torch.save(net.state_dict(), PATH)
```

#### 5.1 重新加载保存的模型

```python
net = Net()
net.load_state_dict(torch.load(PATH))
```

```python
# 好的，现在让我们看看神经网络对以上这些示例的看法：
outputs = net(images)
```

```python
# 输出是 10 类的能量。 一个类别的能量越高，网络就认为该图像属于特定类别。 因此，让我们获取最高能量的指数：
_, predicted = torch.max(outputs, 1)
```

```python
"predicted: "+ " ".join('%5s'% classes[predicted[j]] for j in range(4))
```

```python
correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        
```

```python
# 网络在10000张 test images 上的准去率为
100 * correct/total
```

> 嗯，哪些类的表现良好，哪些类的表现不佳：

```python
class_correct = list(0. for i in range(10))
```

```python
class_total = list(0. for i in range(10))
```

```python
with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = net(images)
        _, predicted = torch.max(outputs, 1)
        c = (predicted == labels).squeeze()
        for i in range(4):
            label = labels[i]
            class_correct[label] += c[i].item()
            class_total[label] += 1
            
```

```python
for i in range(10):
    print('%5s 的准确率 ： %2d %%'%(classes[i], 100*class_correct[i]/class_total[i]))
```
